```python
# 리스트 내 파일 복사
import shutil

def copy_files_by_name(file_names, path_src, path_dst):
    try:
        for file_name in file_names:
            src_file = os.path.join(path_src, file_name)
            dst_file = os.path.join(path_dst, file_name)
            shutil.copyfile(src_file, dst_file)
    except Exception as e:
        print(f"Error occurred: {str(e)}")

!mkdir -p  whl_compared


print("file_names", file_names)
path_src = "/content/whls"
path_dst = "/content/whl_compared"
copy_files_by_name(file_names, path_src, path_dst)
print("\nDone")
```
